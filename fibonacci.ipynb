```python
import fibonacci
import time
```

```python
fib_dict = {}
for n in range(100):
    tic = time.perf_counter()
    fib1 = fibonacci.fib(n)
    toc = time.perf_counter()
    fib2 = fibonacci.fastfib(n)
    toc2 = time.perf_counter()
    print(f"Fibonacci of {n} is {fib1} or {fib2}:\n\tfib {toc - tic:0.4f} seconds\tfastfib {toc2 - toc:0.4f} seconds")
```
